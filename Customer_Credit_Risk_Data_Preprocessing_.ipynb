### 2. What are Tensors?

A tensor is a multi-dimensional array used to represent data.

Types of Tensors:
- Scalar (0D) → single value  
- Vector (1D) → list of values  
- Matrix (2D) → table of values  
- Higher-order Tensor (3D+) → multi-dimensional data  

Tensors are widely used in machine learning and deep learning to store and process data efficiently.

```python
import numpy as np

# 0D Tensor (Scalar)
scalar = np.array(10)
print("Scalar (0D):", scalar)

# 1D Tensor (Vector)
vector = np.array([1, 2, 3, 4])
print("\nVector (1D):", vector)

# 2D Tensor (Matrix)
matrix = np.array([[1, 2], [3, 4]])
print("\nMatrix (2D):\n", matrix)

# 3D Tensor
tensor_3d = np.array([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])
print("\n3D Tensor:\n", tensor_3d)

# Shape of tensors
print("\nShape of 3D Tensor:", tensor_3d.shape)
```

## 4. Handle Missing Values

We fill missing numerical values using mean or median.  
Median is preferred when data has outliers.

```python
mean_imputer = SimpleImputer(strategy='mean')

# Replace directly in original dataframe
df_csv[['age']] = mean_imputer.fit_transform(df_csv[['age']])

print("After Imputation:")
print(df_csv['age'].isnull().sum())

# Check learned value
print("\nMean value learned:", mean_imputer.statistics_)
```

### B. Simple Imputer (Categorical - Most Frequent)

We replace missing categorical values with the most frequent category.

```python
# most_frequent
freq_imputer = SimpleImputer(strategy='most_frequent')

# Only apply on employment_type column (not full df)
df_csv[['employment_type']] = freq_imputer.fit_transform(
    df_csv[['employment_type']]
)

print("\nAfter MOST FREQUENT Imputation:")
print(df_csv['employment_type'])

print("\nMost frequent value learned:", freq_imputer.statistics_)
```

### C. Most Frequent Category Imputation

Most Frequent Category Imputation is used to fill missing values in categorical data.  
In this method, missing values are replaced with the most frequently occurring value (mode) in the column.

This helps maintain the distribution of the data without removing rows.

```python
imputer = SimpleImputer(strategy='most_frequent')

# Apply only on 'gender' column
df_csv[['gender']] = imputer.fit_transform(df_csv[['gender']])

# Check missing values
print("Missing values in gender after imputation:", df_csv['gender'].isnull().sum())

print("Most frequent value:", imputer.statistics_)
```

### C. Missing Indicator + Random Sample

We create a new column indicating missing values and fill missing values using random samples.

```python
# Create Missing Indicator
df_csv['income_missing'] = df_csv['annual_income'].isnull().astype(int)

# Random Sample Imputation
random_sample = df_csv['annual_income'].dropna().sample(
    df_csv['annual_income'].isnull().sum(),
    random_state=42
)

df_csv.loc[df_csv['annual_income'].isnull(), 'annual_income'] = random_sample.values

print(df_csv[['annual_income', 'income_missing']].head())
print("\nMissing values left:", df_csv['annual_income'].isnull().sum())
```

## D. KNN Imputation

KNN Imputer fills missing values using nearest neighbors based on similarity.

```python
cols = ['annual_income', 'loan_amount']

# Apply KNN Imputer
knn = KNNImputer(n_neighbors=5)

df_csv[cols] = knn.fit_transform(df_csv[cols])

# Check missing values
print("Missing values after KNN:\n", df_csv[cols].isnull().sum())
```

### E. MICE Algorithm

MICE (Multiple Imputation by Chained Equations) fills missing values iteratively using other features.

```python
cols = ['credit_score']
# Apply MICE
mice = IterativeImputer(random_state=42)

df_csv[cols] = mice.fit_transform(df_csv[cols])

# Check missing values
print("Missing values after MICE:\n", df_csv[cols].isnull().sum())
```

# Part D: Outlier Handling Techniques

Outliers are extreme values that differ significantly from other observations.  
 They can affect model performance, so handling them is important.


## 1. Z-score Method:
   Detects outliers based on standard deviation.

```python
z = np.abs(stats.zscore(df_csv[['age', 'annual_income', 'loan_amount', 'credit_score']]))

df_z = df_csv[(z < 3).all(axis=1)]

print("Shape after Z-score:", df_z.shape)
```

## 2. IQR Method:
   Uses quartiles to identify extreme values.

```python
Q1 = df_csv[['age','annual_income','loan_amount','credit_score']].quantile(0.25)
Q3 = df_csv[['age','annual_income','loan_amount','credit_score']].quantile(0.75)

IQR = Q3 - Q1

df_iqr = df_csv[~((df_csv[['age','annual_income','loan_amount','credit_score']] < (Q1 - 1.5 * IQR)) |
                  (df_csv[['age','annual_income','loan_amount','credit_score']] > (Q3 + 1.5 * IQR))).any(axis=1)]

print("Shape after IQR:", df_iqr.shape)
```

## 3. Percentile Method:
   Removes values outside a specific percentile range.

```python
lower = df_csv[['annual_income']].quantile(0.01)
upper = df_csv[['annual_income']].quantile(0.99)

df_percentile = df_csv[(df_csv['annual_income'] >= lower.values[0]) &
                       (df_csv['annual_income'] <= upper.values[0])]

print("Shape after Percentile:", df_percentile.shape)
```

## 4. Winsorization:
   Caps extreme values instead of removing them.

```python
df_csv['annual_income_winsor'] = winsorize(df_csv['annual_income'], limits=[0.01, 0.01])

print(df_csv[['annual_income', 'annual_income_winsor']].head())
```

```python
plt.figure(figsize=(12,5))

# Before
plt.subplot(1,2,1)
sns.histplot(df_csv['annual_income'], kde=True)
plt.title("Before Winsorization")

# After
plt.subplot(1,2,2)
sns.histplot(df_csv['annual_income_winsor'], kde=True)
plt.title("After Winsorization")

plt.show()
```

## 9. Encoding Categorical Variables

Categorical variables need to be converted into numerical format for machine learning models.

   Applied on education_level (ordered categories)
   Example: Primary < Secondary < Graduate < Post-Graduate


```python
edu_order = [['Primary', 'Secondary', 'Graduate', 'Post-Graduate']]

# Apply encoder
ordinal_encoder = OrdinalEncoder(categories=edu_order)

df_json['education_level_enc'] = ordinal_encoder.fit_transform(
    df_json[['education_level']]
)

df_json['education_level_enc'] = df_json['education_level_enc'].astype(int)

print("Ordinal Encoding Applied:")
print(df_json[['education_level', 'education_level_enc']].head())
```

### 2. Label Encoding:
   Used for binary or simple categorical variables.
   Example: gender → Male = 1, Female = 0


```python
label_encoder = LabelEncoder()

df_csv['gender_enc'] = label_encoder.fit_transform(df_csv['gender'])

print("\nLabel Encoding Applied:")
print(df_csv[['gender', 'gender_enc']].head())
```

### 3. One-Hot Encoding:
   Used for nominal categories where no order exists.
   Example: region, loan_purpose → converted into multiple columns

```python
df_ohe = pd.get_dummies(df_csv, columns=['region', 'loan_purpose'], drop_first=True)

print("\nOne-Hot Encoding Applied:")
print(df_ohe[['region_North', 'region_South', 'region_West', 'loan_purpose_Car', 'loan_purpose_Education', 'loan_purpose_Home', 'loan_purpose_Other']].head().to_string())
```

## 10. Encoding Numerical Features

Numerical feature encoding is the process of converting continuous numerical values into discrete categories or groups.

This helps:
- Handle outliers
- Capture non-linear relationships
- Improve model interpretability

### 1. Binning:
   Continuous values are divided into groups (bins) to simplify data and reduce noise.


```python
df_csv['income_bin'] = pd.cut(
    df_csv['annual_income'],
    bins=3,
    labels=['Low', 'Medium', 'High']
)

df_csv['repayment_bin'] = pd.cut(
    df_csv['repayment_history'],
    bins=3,
    labels=['Poor', 'Average', 'Good']
)

print(df_csv[['income_bin', 'annual_income', 'repayment_bin','repayment_history']].head())
```

### 2. Binarization:
   Binarization converts numerical values into binary (0 or 1) based on a threshold.

```python
binarizer = Binarizer(threshold=700)

df_csv['credit_binary'] = binarizer.fit_transform(
    df_csv[['credit_score']]
)

print(df_csv[['credit_score', 'credit_binary']].head())
```

### 3. Quantile Binning:
   Divides data into equal-sized groups (equal frequency bins).

```python
df_csv['transaction_quantile'] = pd.qcut(
    df_csv['transaction_count'],
    q=4,
    labels=['Q1', 'Q2', 'Q3', 'Q4']
)

print(df_csv[['transaction_count', 'transaction_quantile']].head())
```

### 4. K-Means Binning:
   Uses clustering to group similar numerical values.

```python
kbin = KBinsDiscretizer(
    n_bins=3,
    encode='ordinal',
    strategy='kmeans'
)

df_csv[['transaction_kmeans']] = kbin.fit_transform(
    df_csv[['transaction_count']]
)

print(df_csv[['transaction_count', 'transaction_kmeans']].head())
```

# Part F: Feature Scaling

## 11. Multiple Scaling Methods:

### 1. Standardization (Z-score Scaling)

Standardization transforms data such that:
- Mean = 0  
- Standard Deviation = 1  

This is the most commonly used scaling technique.

```python
scaler = StandardScaler()

df_csv[['annual_income_std', 'loan_amount_std']] = scaler.fit_transform(
    df_csv[['annual_income', 'loan_amount']]
)

print(df_csv[['annual_income', 'annual_income_std']].head())
```

### 2. Normalization

Normalization scales values to unit norm (L1 or L2).

It is useful when direction of data matters more than magnitude.

#### I. Min-Max Scaling

Min-Max scaling transforms data into a fixed range [0,1].

It preserves the shape of distribution but rescales values.

```python
minmax = MinMaxScaler()

df_csv[['annual_income_minmax', 'loan_amount_minmax']] = minmax.fit_transform(
    df_csv[['annual_income', 'loan_amount']]
)

print(df_csv[['annual_income_minmax', 'loan_amount_minmax']].head())
```

#### II. MaxAbs Scaling

MaxAbs scaling scales data based on maximum absolute value.

It is useful for sparse data and keeps zero values intact.

```python
maxabs = MaxAbsScaler()

df_csv[['annual_income_maxabs', 'loan_amount_maxabs']] = maxabs.fit_transform(
    df_csv[['annual_income', 'loan_amount']]
)

print(df_csv[['annual_income_maxabs', 'loan_amount_maxabs']].head())
```

#### III. Robust Scaling

Robust scaling uses median and IQR (Interquartile Range).

It is resistant to outliers and works well when data contains extreme values.

```python
robust = RobustScaler()

df_csv[['annual_income_robust', 'loan_amount_robust']] = robust.fit_transform(
    df_csv[['annual_income', 'loan_amount']]
)

print(df_csv[['annual_income_robust', 'loan_amount_robust']].head())
```

# Part G: Feature Construction & Transformation

## 12. Transformations:

### I. FunctionTransformer

FunctionTransformer is used to apply mathematical transformations like:
- Log
- Reciprocal
- Square Root

Applied on:
- spending_ratio

```python
# Log Transform
log_transform = FunctionTransformer(np.log1p)
df_csv['spending_log'] = log_transform.fit_transform(df_csv[['spending_ratio']])

# Square Root Transform
sqrt_transform = FunctionTransformer(np.sqrt)
df_csv['spending_sqrt'] = sqrt_transform.fit_transform(df_csv[['spending_ratio']])

# Reciprocal Transform
df_csv['spending_reciprocal'] = 1 / (df_csv['spending_ratio'] + 1)

print(df_csv[['spending_ratio', 'spending_log', 'spending_sqrt']].head())
```

### II. PowerTransformer

PowerTransformer is used to make data more normally distributed.

Methods:
- Box-Cox (only positive data)
- Yeo-Johnson (works with negative values)

Applied on:
- loan_amount
- annual_income

```python
# Box-Cox (only for positive values)

pt_boxcox = PowerTransformer(method='box-cox')

df_csv[['loan_boxcox', 'income_boxcox']] = pt_boxcox.fit_transform(
    df_csv[['loan_amount', 'annual_income']]
)

print(df_csv[['loan_amount', 'loan_boxcox']].head())

```

```python
# Yeo-Johnson (safe for all values)
pt = PowerTransformer(method='yeo-johnson')

df_csv[['loan_power', 'income_power']] = pt.fit_transform(
    df_csv[['loan_amount', 'annual_income']]
)

print(df_csv[['loan_amount', 'loan_power']].head())
```

### II. Column Transformer

ColumnTransformer is used to apply different preprocessing steps to different columns.

- Numerical features → scaling
- Categorical features → encoding

This helps build a complete preprocessing pipeline.

```python
num_cols = ['annual_income', 'loan_amount']
cat_cols = ['region', 'loan_purpose']

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_cols),
        ('cat', OneHotEncoder(drop='first'), cat_cols)
    ]
)

df_transformed = preprocessor.fit_transform(df_csv)

# Column names generate karo
ohe_cols = preprocessor.named_transformers_['cat'].get_feature_names_out(cat_cols)

all_cols = list(num_cols) + list(ohe_cols)

# DataFrame ma convert karo
df_final = pd.DataFrame(df_transformed, columns=all_cols)

# Output dekhao
print(df_final.head().to_string())
```
